### Calculate Avg Freq

```python
data_stats = pd.read_csv('../test_align_freq.tsv', sep='\t', header=None, names=['rel_kb', 'rel_oie', 'freq'])
```

```python
data_stats
```

```python
data_stats_avg = data_stats.groupby(['rel_kb', 'rel_oie'], as_index=False).mean().groupby('rel_kb')['freq'].mean()
```

```python
data_stats_avg
```

```python
data_stats_avg.to_csv('../test_align_freq_avg.tsv', sep='\t')
```
