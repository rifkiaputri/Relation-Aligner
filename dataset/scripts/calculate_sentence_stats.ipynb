```python
import pandas as pd
```

### Sentence Stats

```python
sent_data = pd.read_csv('../test_sentences_filtered.tsv', sep='\t', header=None, names=['e1_id', 'e2_id', 'rel_id', 'e1_label', 'e2_label', 'rel_label', 'sent'], quoting=3)
```

```python
len(sent_data)
```

```python
stats = sent_data.groupby(['e1_id', 'e2_id', 'rel_id']).size()
```

```python
stats
```

```python
stats.to_csv('../stats/train_filtered_stats.tsv', sep='\t')
```

### Open IE Rel Stats

```python
headers = ['e1_kb', 'rel_kb', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie', 'rel_id', 'label', 'e1_oie_root', 'e2_oie_root', 'e1_desc', 'e2_desc']
train_align = pd.read_csv('../train_align_filtered.tsv', sep='\t', header=None, names=headers)
valid_align = pd.read_csv('../valid_align_filtered.tsv', sep='\t', header=None, names=headers)
test_align = pd.read_csv('../test_align_filtered_balanced.tsv', sep='\t', header=None, names=headers)
```

```python
align_all = pd.concat([train_align, valid_align, test_align])
```

```python
align_all
```

```python
oie_rel_stats = align_all.groupby(['rel_oie']).size()
```

```python
oie_rel_stats.to_csv('../stats/oie_rels_filtered.tsv', sep='\t')
```

```python
kb_rel_stats = align_all.groupby(['rel_kb']).size()
```

```python
kb_rel_stats
```

```python
kb_rel_stats.to_csv('../stats/kb_rels_filtered.tsv', sep='\t')
```

```python
oie_triples = align_all.filter(['e1_oie', 'rel_oie', 'e2_oie'], axis=1)
oie_triples = oie_triples.drop_duplicates()
```

```python
oie_triples
```

```python
oie_triples.to_csv('../oie_triples.tsv', sep='\t', header=False, index=False)
```
