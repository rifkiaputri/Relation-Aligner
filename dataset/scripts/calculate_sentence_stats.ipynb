```python
import pandas as pd
```

### Sentence Stats

```python
sent_data = pd.read_csv('../train_sentences_filtered.tsv', sep='\t', header=None, names=['e1_id', 'e2_id', 'rel_id', 'e1_label', 'e2_label', 'rel_label', 'sent'])
```

```python
sent_data
```

```python
stats = sent_data.groupby(['e1_id', 'e2_id', 'rel_id']).size()
```

```python
stats.to_csv('../stats/train_stats.tsv', sep='\t')
```

### Open IE Rel Stats

```python
train_align = pd.read_csv('../train_align.tsv', sep='\t', header=None, names=['e1_kb', 'rel_kb', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie', 'rel_id', 'label'])
valid_align = pd.read_csv('../valid_align.tsv', sep='\t', header=None, names=['e1_kb', 'rel_kb', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie', 'rel_id', 'label'])
test_align = pd.read_csv('../test_align.tsv', sep='\t', header=None, names=['e1_kb', 'rel_kb', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie', 'rel_id', 'label'])
```

```python
align_all = pd.concat([train_align, valid_align, test_align])
```

```python
align_all
```

```python
oie_rel_stats = align_all.groupby(['rel_oie']).size()
```

```python
oie_rel_stats.to_csv('../stats/oie_rels.tsv', sep='\t')
```

```python
kb_rel_stats = align_all.groupby(['rel_kb']).size()
```

```python
kb_rel_stats.to_csv('../stats/kb_rels.tsv', sep='\t')
```
