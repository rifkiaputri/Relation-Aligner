```python
import json
```

```python
with open('../../predict/cesi/cluster_rel.json') as f:
    rel_data = json.load(f)
```

```python
cluster = rel_data['cluster']
```

```python
rel_to_id = {}
for i in range(len(cluster)):
    for rel in cluster[i]:
        rel_to_id[rel] = i
```

```python
rel_to_id
```

```python
with open('../../predict/cesi/rel_to_cluster_id.json', 'w') as outfile:
    json.dump(rel_to_id, outfile)
```
