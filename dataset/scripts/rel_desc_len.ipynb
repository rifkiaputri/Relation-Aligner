```python
import pandas as pd
import sys
sys.path.append('../../')
import moses
tokenizer = moses.MosesTokenizer()
```

```python
import re
```

```python
kb_rels = pd.read_csv('../relation_desc.tsv', sep='\t', header=None, names=['id', 'label', 'desc'])
kb_rels['len_desc'] = 0
```

```python
kb_rels
```

```python
for i in range(len(kb_rels)):
    text = ' '.join(re.findall(r'\w+', kb_rels['desc'][i], flags=re.UNICODE)).lower()
    tokens = tokenizer.tokenize(text)
    kb_rels['len_desc'][i] = len(tokens)
```

```python
kb_rels
```

```python
kb_rels.to_csv('../relation_desc_len.tsv', sep='\t', header=False, index=False)
```
