```python
import pandas as pd
```

```python
headers = ['e1_kb', 'rel_kb', 'rel_id', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie',
           'e1_kb_id', 'e2_kb_id', 'e1_oie_id', 'e2_oie_id',
           'e1_oie_root', 'e2_oie_root', 'label']
data_align = pd.read_csv('../train_align_filtered_all.tsv', sep='\t', header=None, names=headers)
```

```python
data_align
```

```python
stats_label = data_align.groupby(['label']).size()
num_pos, num_neg = stats_label[0], stats_label[1]
```

```python
num_pos, num_neg
```

```python
remove_idxs = []
triples_max = 70
i, num_triples = 0, 0
prev_e1, prev_rel, prev_e2, prev_rel_oie, prev_label = data_align['e1_kb'][0], data_align['rel_kb'][0], data_align['e2_kb'][0], data_align['rel_oie'][0], data_align['label'][0]
len_data = len(data_align)
len_remove = num_pos - num_neg

while i < len_data or len(remove_idxs) < len_remove:
    e1, rel, e2, rel_oie, label = data_align['e1_kb'][i], data_align['rel_kb'][i], data_align['e2_kb'][i], data_align['rel_oie'][i], data_align['label'][i]
    if prev_e1 == e1 and prev_rel == rel and prev_e2 == e2 and prev_label == label:
        num_triples += 1
    else:
        num_triples = 1
    
    label = data_align['label'][i]
    if num_triples > triples_max:
        remove_idxs.append(i)
    
    i += 1
    prev_e1, prev_rel, prev_e2, prev_label = e1, rel, e2, label
```

```python
len(remove_idxs)
```

```python
data_align.drop(data_align.index[remove_idxs], inplace=True)
```
