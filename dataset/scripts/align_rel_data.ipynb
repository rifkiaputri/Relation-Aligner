```python
import pandas as pd
import spacy
nlp = spacy.load('en_core_web_sm')
from wikidata.client import Client
client = Client()
```

```python
import time
```

```python
data_kb = pd.read_csv('../test_sentences_filtered.tsv', sep='\t', header=None, names=['e1_id', 'e2_id', 'rel_id', 'e1_label', 'e2_label', 'rel_label', 'sent'])
data_openie = pd.read_csv('../test_openie_filtered.tsv', sep='\t', header=None, names=['has_kb_rel', 'sent_id', 'e1', 'rel', 'e2'])
```

```python
def get_noun_root(text):
    doc = nlp(text)
    root = 'N/A'
    for chunk in doc.noun_chunks:
        if chunk.root.dep_ == 'ROOT':
            root = chunk.root.text
    return root
```

```python
dict_desc = {}

def get_entity_description(ent_id):
    try:
        desc = dict_desc[ent_id] 
    except KeyError:
        entity = client.get(ent_id, load=True)
        desc = entity.description
        dict_desc[ent_id] = desc
    return desc
```

```python
data_kb['e1_desc'] = 'N/A'
data_kb['e2_desc'] = 'N/A'
len_data_kb = len(data_kb['e1_id'])
for i in range(len_data_kb):
    print(i, 'of', len_data_kb - 1)
    try:
        e1_desc = get_entity_description(data_kb['e1_id'][i])
        e2_desc = get_entity_description(data_kb['e2_id'][i])
        data_kb['e1_desc'][i] = e1_desc
        data_kb['e2_desc'][i] = e2_desc
    except OSError:
        print('OSError')
        time.sleep(1)
        e1_desc = get_entity_description(data_kb['e1_id'][i])
        e2_desc = get_entity_description(data_kb['e2_id'][i])
        data_kb['e1_desc'][i] = e1_desc
        data_kb['e2_desc'][i] = e2_desc
    except URLError:
        print('URLError')
        time.sleep(1)
        e1_desc = get_entity_description(data_kb['e1_id'][i])
        e2_desc = get_entity_description(data_kb['e2_id'][i])
        data_kb['e1_desc'][i] = e1_desc
        data_kb['e2_desc'][i] = e2_desc
    
```

```python
align_data = []
len_openie = len(data_openie['e1'])
for i in range(len_openie):
    #print(i, 'of', len_openie - 1)
    sent_idx = data_openie['sent_id'][i]
    if data_openie['has_kb_rel'][i] == 1:
        # Extract noun chunker
        e1_kb = data_kb['e1_label'][sent_idx]
        e2_kb = data_kb['e2_label'][sent_idx]
        e1_oie = data_openie['e1'][i]
        e2_oie = data_openie['e2'][i]
        e1_root = get_noun_root(e1_oie)
        e2_root = get_noun_root(e2_oie)
        e1_desc = data_kb['e1_desc'][sent_idx]
        e2_desc = data_kb['e2_desc'][sent_idx]
        
        if ((e1_root in e1_kb) or (e1_kb in e1_root) or (e1_kb in e1_oie) or (e1_oie in e1_kb)) and ((e2_root in e2_kb) or (e2_kb in e2_root)):
            # same class: 0
            align_data.append((e1_kb, data_kb['rel_label'][sent_idx], e2_kb,
                               e1_oie, data_openie['rel'][i], e2_oie,
                               data_kb['rel_id'][sent_idx], '0', e1_root, e2_root, e1_desc, e2_desc))
        else:
            align_data.append((e1_kb, data_kb['rel_label'][sent_idx], e2_kb,
                               e1_oie, data_openie['rel'][i], e2_oie,
                               data_kb['rel_id'][sent_idx], '1', e1_root, e2_root, e1_desc, e2_desc))
```

```python
with open('../test_align_filtered.tsv', 'w') as f:
    for item in align_data:
        len_item = len(item) - 1
        for i in range(len_item):
            f.write(str(item[i]) + '\t')
        f.write(str(item[len_item]) + '\n')
f.closed
```
