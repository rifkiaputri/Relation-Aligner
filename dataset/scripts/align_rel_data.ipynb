```python
def get_noun_root(text, basic_root=True):
    doc = nlp(text)
    root = 'N/A'
    for chunk in doc.noun_chunks:
        if chunk.root.dep_ == 'ROOT':
            if basic_root == True:
                root = chunk.root.text
            else:
                root = chunk.text
    return root
```

```python
def get_entity_tuple_id(e1, rel, e2, e1_kb, e2_kb):
    e1_wiki_id = ''
    e2_wiki_id = ''
    e1_label = ''
    e2_label = ''
    ent_result, ent_id_dict = enteater.get_entity(e1 + ' ' + rel + ' ' + e2, detect_property=False)
    ent_label_dict = dict((v,k) for k,v in ent_id_dict.items())
    
    # Check e1
    if e1_kb in ent_id_dict.values():
        e1_wiki_id = e1_kb
        e1_label = ent_label_dict[e1_kb]
        
    # Check e2
    if e2_kb in ent_id_dict.values():
        e2_wiki_id = e2_kb
        e2_label = ent_label_dict[e2_kb]
    
    return (e1_wiki_id, e1_label, e2_wiki_id, e2_label)
```

```python
align_data = []
len_openie = len(data_openie['e1'])
for i in range(len_openie):
    print(i, 'of', len_openie - 1)
    sent_idx = data_openie['sent_id'][i]
    e1_kb_id = data_kb['e1_id'][sent_idx]
    e2_kb_id = data_kb['e2_id'][sent_idx]
    e1_kb = data_kb['e1_label'][sent_idx]
    e2_kb = data_kb['e2_label'][sent_idx]
    rel_kb = data_kb['rel_label'][sent_idx]
    rel_kb_id = data_kb['rel_id'][sent_idx]
    e1_oie = data_openie['e1'][i]
    e2_oie = data_openie['e2'][i]
    rel_oie = data_openie['rel'][i]
    e1_oie_id, e1_oie_root, e2_oie_id, e2_oie_root = get_entity_tuple_id(e1_oie, rel_oie, e2_oie, e1_kb_id, e2_kb_id)
    
    if (e1_oie_id == e1_kb_id or e1_kb == e1_oie) and (e2_oie_id == e2_kb_id or e2_kb == e2_oie):
        # same class (similar relation): 0
        align_data.append((e1_kb, rel_kb, rel_kb_id, e2_kb,
                           e1_oie, rel_oie, e2_oie,
                           e1_kb_id, e2_kb_id, e1_oie_id, e2_oie_id, e1_oie_root, e2_oie_root,
                           '0'))
    elif (e1_oie_id == e1_kb_id or e1_kb == e1_oie) or (e2_oie_id == e2_kb_id or e2_kb == e2_oie):
        align_data.append((e1_kb, rel_kb, rel_kb_id, e2_kb,
                           e1_oie, rel_oie, e2_oie,
                           e1_kb_id, e2_kb_id, e1_oie_id, e2_oie_id, e1_oie_root, e2_oie_root,
                           '1'))
```

```python
with open('../test_align_filtered.tsv', 'w') as f:
    for item in align_data:
        len_item = len(item) - 1
        for i in range(len_item):
            f.write(str(item[i]) + '\t')
        f.write(str(item[len_item]) + '\n')
f.closed
```

```python
align_df_pos = align_df[(align_df.label != 1) & (align_df.rel_kb_id != 'P31') & (align_df.rel_kb_id != 'P17')
                       & (align_df.rel_kb_id != 'P131')]
```

```python
align_df_pos
```

```python
align_data = []
len_pos = len(align_df_pos)
for idx, row in align_df_pos.iterrows():
    print(idx, 'of', len_pos)
    e1, e1_id, rel, rel_id, e2, e2_id = row['e1_kb'], row['e1_kb_id'], row['rel_kb'], row['rel_kb_id'], row['e2_kb'], row['e2_kb_id']
    pos_list = align_df_pos[(align_df_pos.rel_kb_id == rel_id) & (align_df_pos.e1_kb_id != e1_id) & (align_df_pos.e2_kb_id != e2_id)]
    if len(pos_list) > 0:
        for idx_l, row_l in pos_list.iterrows():
            align_data.append({
                'e1_kb': e1,
                'rel_kb': rel,
                'rel_kb_id': rel_id,
                'e2_kb': e2,
                'e1_oie': row_l['e1_oie'],
                'rel_oie': row_l['rel_oie'],
                'e2_oie': row_l['e2_oie'],
                'e1_kb_id': e1_id,
                'e2_kb_id': e2_id,
                'e1_oie_id': row_l['e1_oie_id'],
                'e2_oie_id': row_l['e2_oie_id'],
                'e1_oie_root': row_l['e1_oie_root'],
                'e2_oie_root': row_l['e2_oie_root'],
                'label': '0'
            })
```

```python
align_df = align_df.append(align_data)
```
