```python
import pandas as pd
```

```python
data_kb = pd.read_csv('dataset/train_sentences.tsv', sep='\t', header=None, names=['e1_id', 'e2_id', 'rel_id', 'e1_label', 'e2_label', 'rel_label', 'sent'])
data_openie = pd.read_csv('dataset/train_openie.tsv', sep='\t', header=None, names=['has_kb_rel', 'sent_id', 'e1', 'rel', 'e2'])
```

```python
data_kb
```

```python
data_openie
```

```python
align_data = []
len_openie = len(data_openie['e1'])
for i in range(len_openie):
    #print(i, 'of', len_openie - 1)
    sent_idx = data_openie['sent_id'][i]
    if data_openie['has_kb_rel'][i] == 1:
        # same class: 0
        align_data.append((data_kb['e1_label'][sent_idx], data_kb['rel_label'][sent_idx], data_kb['e2_label'][sent_idx],
                           data_openie['e1'][i], data_openie['rel'][i], data_openie['e2'][i],
                           data_kb['rel_id'][sent_idx], '0'))
    else:
        align_data.append((data_kb['e1_label'][sent_idx], data_kb['rel_label'][sent_idx], data_kb['e2_label'][sent_idx],
                           data_openie['e1'][i], data_openie['rel'][i], data_openie['e2'][i],
                           data_kb['rel_id'][sent_idx], '1'))
```

```python
with open('dataset/train_align.tsv', 'w') as f:
    for item in align_data:
        len_item = len(item) - 1
        for i in range(len_item):
            f.write(item[i] + '\t')
        f.write(item[len_item] + '\n')
f.closed
```
