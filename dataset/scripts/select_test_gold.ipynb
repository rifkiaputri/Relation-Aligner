```python
import pandas as pd
```

```python
headers = ['e1_kb', 'rel_kb', 'rel_id', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie',
           'e1_kb_id', 'e2_kb_id', 'e1_oie_id', 'e2_oie_id',
           'e1_oie_root', 'e2_oie_root', 'label']
test_all = pd.read_csv('../test_align_all.tsv', sep='\t', header=None, names=headers, quoting=3)
test_balanced = pd.read_csv('../test_align_all_balanced.tsv', sep='\t', header=None, names=headers, quoting=3)
```

```python
len(test_balanced), len(test_all)
```

```python
test = test_all.merge(test_balanced,on=['e1_kb', 'rel_kb', 'rel_id', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie',
           'e1_kb_id', 'e2_kb_id', 'e1_oie_id', 'e2_oie_id',
           'e1_oie_root', 'e2_oie_root'],how='left')
```

```python
test
```

```python
test = test[test.label_y.isnull()]
test = test.drop('label_y', 1)
```

```python
test
```

```python
test.to_csv('../test_gold_candidates.tsv', sep='\t', index=False)
```

```python
test = test.drop_duplicates(subset=['e1_kb', 'rel_kb', 'rel_id', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie'], keep="first")
```
