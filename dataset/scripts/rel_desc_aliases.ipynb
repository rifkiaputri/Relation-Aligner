```python
import pandas as pd
```

```python
kb_rels = pd.read_csv('../relation_desc_len.tsv', sep='\t', header=None, names=['id', 'label', 'desc', 'len_desc'])
kb_rels['aliases'] = ''
```

```python
kb_rels
```

```python
rels_data = pd.read_csv('../relation_data.tsv', sep='\t')
```

```python
rels_data
```

```python
id_to_aliases = {}
for i in range(len(rels_data)):
    id_to_aliases[rels_data['id'][i]] = rels_data['aliases'][i]
```

```python
id_to_aliases['P31']
```

```python
for i in range(len(kb_rels)):
    kb_rels['aliases'][i] = id_to_aliases[kb_rels['id'][i]]
```

```python
len(kb_rels[pd.isnull(kb_rels['aliases'])])
```
