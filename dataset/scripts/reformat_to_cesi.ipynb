```python
import pandas as pd
import io, json
import codecs
```

```python
test = pd.read_csv('../test_align_gold.tsv', sep='\t', header=None, names=['e1_kb', 'rel_kb', 'rel_kb_id', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie', 'e1_kb_id', 'e2_kb_id', 'e1_oie_id', 'e2_oie_id', 'e1_oie_root', 'e2_oie_root', 'label'], encoding='utf8')
```

```python
test
```

```python
triples = []
for i in range(len(test)):
    e1_oie = test['e1_oie'][i].encode('ascii', 'ignore').decode('ascii')
    rel_oie = test['rel_oie'][i].encode('ascii', 'ignore').decode('ascii')
    e2_oie = test['e2_oie'][i].encode('ascii', 'ignore').decode('ascii')
    item_oie = {
        "_id": i,
        "triple": [
            e1_oie,
            rel_oie,
            e2_oie
        ],
        "triple_norm": [
            e1_oie.lower(),
            rel_oie.lower(),
            e2_oie.lower()
        ],
        "true_link": {
            "subject": test['e1_kb_id'][i],
            "object":  test['e2_kb_id'][i]
        },
        "src_sentences": [
            e1_oie + ' ' + rel_oie + ' ' + e2_oie
        ],
        "entity_linking": {
            "subject":  test['e1_kb'][i],
            "object":   test['e2_kb'][i]
        },
        "kbp_info": []
    }
    
    e1_kb = test['e1_kb'][i].encode('ascii', 'ignore').decode('ascii')
    rel_kb = test['rel_kb'][i].encode('ascii', 'ignore').decode('ascii')
    e2_kb = test['e2_kb'][i].encode('ascii', 'ignore').decode('ascii')
    item_kb = {
        "_id": i,
        "triple": [
            e1_kb,
            rel_kb,
            e2_kb
        ],
        "triple_norm": [
            e1_kb.lower(),
            rel_kb.lower(),
            e2_kb.lower()
        ],
        "true_link": {
            "subject": test['e1_kb_id'][i],
            "object":  test['e2_kb_id'][i]
        },
        "src_sentences": [
            e1_kb + ' ' + rel_kb + ' ' + e2_kb
        ],
        "entity_linking": {
            "subject":  test['e1_kb'][i],
            "object":   test['e2_kb'][i]
        },
        "kbp_info": []
    }
    triples.append(item_kb)
    triples.append(item_oie)
```

```python
with codecs.open('../../../cesi/data/reverb45k/reverb45k_test', 'w', encoding='utf-8') as f:
    for item in triples:
        f.write(json.dumps(item, ensure_ascii=False))
        f.write('\n')
```
