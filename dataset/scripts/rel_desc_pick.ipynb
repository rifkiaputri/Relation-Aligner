```python
import pandas as pd
```

```python
rel_data = pd.read_csv('../relation_data.tsv', sep='\t', index_col=0, encoding='utf-8')
```

```python
rel_data
```

```python
# Read occurred relation list
rel = pd.DataFrame()
list_ = []
for file_ in ['../train_rel_kb_freq.tsv', '../valid_rel_kb_freq.tsv', '../test_rel_kb_freq.tsv']:
    df = pd.read_csv(file_, sep='\t', header=None, names=['id', 'num'])
    list_.append(df)
rel = pd.concat(list_)
```

```python
rel
```

```python
rel = rel.drop('num', 1)
rel = rel.drop_duplicates(keep='first', inplace=False)
rel = rel.reset_index(drop=True)
rel['label'] = ''
rel['desc'] = ''
```

```python
for i in range(len(rel)):
    rel_id = rel['id'][i]
    rel['label'][i] = rel_data['label'][rel_id]
    rel['desc'][i] = rel_data['description'][rel_id].replace('"', '')
```

```python
rel
```

```python
rel.to_csv('../relation_desc.tsv', header=None, index=None, sep='\t')
```
