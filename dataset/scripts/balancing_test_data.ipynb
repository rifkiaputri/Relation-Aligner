```python
import pandas as pd
import csv
```

```python
headers = ['e1_kb', 'rel_kb', 'rel_id', 'e2_kb', 'e1_oie', 'rel_oie', 'e2_oie',
           'e1_kb_id', 'e2_kb_id', 'e1_oie_id', 'e2_oie_id',
           'e1_oie_root', 'e2_oie_root', 'label']
data_align = pd.read_csv('../test_align_filtered_all.tsv', sep='\t', header=None, names=headers, quoting=csv.QUOTE_NONE)
```

```python
data_align
```

```python
stats_label = data_align.groupby(['label']).size()
num_pos, num_neg = stats_label[0], stats_label[1]
```

```python
num_pos, num_neg
```

```python
remove_idxs = []
num_remove = num_pos - num_neg
triples_max = 5
i, num_triples = 0, 0
prev_e1, prev_rel, prev_e2 = data_align['e1_kb'][0], data_align['rel_kb'][0], data_align['e2_kb'][0]

while len(remove_idxs) < num_remove:
    e1, rel, e2 = data_align['e1_kb'][i], data_align['rel_kb'][i], data_align['e2_kb'][i]
    if prev_e1 == e1 and prev_rel == rel and prev_e2 == e2:
        num_triples += 1
    else:
        num_triples = 1
    
    label = data_align['label'][i]
    if label == 0 and num_triples > triples_max:
        remove_idxs.append(i)
    
    i += 1
    prev_e1, prev_rel, prev_e2 = e1, rel, e2
```

```python
len(remove_idxs)
```

```python
data_align.drop(data_align.index[remove_idxs], inplace=True)
```
